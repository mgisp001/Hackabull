```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
from pandas_datareader.data import DataReader
from datetime import datetime
from __future__ import division
```

```python
end = datetime.now()
start = datetime(end.year-1,end.month,end.day)
```

```python
baseline_list = ['ARCO','MFA','SRCI']
```

```python
for stock in baseline_list:
    globals()[stock]= DataReader(stock,'yahoo',start,end)
```

```python
ARCO.describe()
```

```python
closing_df_baseline = DataReader(baseline_list,'yahoo',start,end)['Adj Close']
```

```python
closing_df_baseline.head()
```

```python
baseline_returns = closing_df_baseline.pct_change()

baseline_returns.head()
```

```python
comparative_stock_list = ['AGFS','POPE','ARMK','INST','STC','TTM']
```

```python
for stock in comparative_stock_list:
    globals()[stock]= DataReader(stock,'yahoo',start,end)
```

```python
POPE.head()
```

```python
closing_df_comparative = DataReader(comparative_stock_list,'yahoo',start,end)['Adj Close']
```

```python
closing_df_comparative.head()
```

```python
comparative_returns = closing_df_comparative.pct_change()

comparative_returns.head()
```

```python
total_stock = ['AGFS','POPE','ARMK','INST','STC','TTM',
               'ARCO','MFA','SRCI']

for stock in total_stock:
    globals()[stock]= DataReader(stock,'yahoo',start,end)
    
closing_df_totalstock = DataReader(total_stock,'yahoo',start,end)
['Adj Close']

total_returns = closing_df_totalstock.pct_change()

total_returns.head()
```

```python
sns.jointplot('POPE','AGFS',comparative_returns,kind = 'scatter', color='green')
```
